```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
# Load data
data = pd.read_csv("1. Weather Data.csv")
data.head()
```

```python
# Check the info
data.info()
```

```python
# Convert the data column to date
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data.info()
```

### Questions

```python
# Q1: Records where Weather == Clear
data[data.Weather == 'Clear']
```

```python
# Q2: Number of records with wind speed == 4km/hr
data[data['Wind Speed_km/h'] == 4].shape[0]
```

```python
# Q3: Check for null values
data.isnull().sum()
```

```python
# Q4: Rename Weather column to Weather_Condition
data.rename(columns = {'Weather':'Weather_Condition'}, inplace=True)
data.tail(3)
```

```python
# Q5: Mean visibility
data.Visibility_km.mean()
```

```python
# Q6: records where wind speed > 24 km/hr and visibility == 25 km
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]
```

```python
# Q7: Mean value for each weather condition
data[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 
'Wind Speed_km/h', 'Visibility_km', 'Press_kPa'
]].mean()
```

```python
# Q8: where Weather == Clear and the relative humidity > 50, or visibility > 40
data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]
```

```python
# Q9: Number of weather conditions that include snow
data[data['Weather_Condition'].str.contains('Snow', case = False)].shape[0]
```
